# cloud-inference-platform quick walkthrough

This notebook creates an in-memory router, records latency samples, and exercises the FastAPI app with `TestClient`.

```python
import fakeredis
from fastapi.testclient import TestClient

from router.slo_router import SLOConfig, SLORouter, app
```

```python
router = SLORouter(SLOConfig(p95_target_ms=120.0, degradation_threshold=0.75), fakeredis.FakeRedis())
for latency in range(1, 101):
    router.record_latency(float(latency))
router._get_current_p95_latency()
```
